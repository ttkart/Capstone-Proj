```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_csv('merge.csv')

df.info()
```

```python
sns.set(rc={'figure.figsize':(20,8)})
hm = sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
```

```python
# make new field 'Total_precepitation': precipitation sum of two states
df['Total_precepitation'] = df['Monthly Total Precepitation_depart'] + df['Monthly Total Precepitation_arrival']
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style('whitegrid')
sns.scatterplot(data=df, x='Total_precepitation', y='Total_Mkt_Fare', s=100, color='.1', marker='+')
```

```python
# make new field 'Average_increment'
df['Average_increment'] = (df.Carrier_MinFareIncrement + df.CarrierMaxFareIncrement)/2
```

```python
df.head()
```

```python
palette = sns.color_palette("mako_r", 4)
sns.lineplot(data=df, x='Year', y='Average_increment', hue='Quarter', errorbar=None, palette=palette)
```

```python
depart = df.groupby(['depart_city']).mean()
Average_Mkt_Fare_depart = depart.sort_values('Total_Mkt_Fare', ascending=False) \
    [['Total_Mkt_Fare']]
Average_Mkt_Fare_depart['id'] = Average_Mkt_Fare_depart.index

Average_Mkt_Fare_depart
```

```python
citi = pd.read_csv('cities.csv')

new_df = Average_Mkt_Fare_depart.merge(citi, how='inner', on='id')

new_df.iloc[:10]
```

```python
top10_depart = new_df.iloc[:10]
ax = sns.barplot(x='city', y='Total_Mkt_Fare', data=top10_depart, color='blue')

bars = ax.containers[0]
ax.bar_label(bars, labels=[f'${value:.2f}' for value in bars.datavalues])

ax.set_title('Top 10 Average Total Market Fare by Departure City')

plt.show()
```

```python
arrive = df.groupby(['arrival_city']).mean()
Average_Mkt_Fare_arrive = arrive.sort_values('Total_Mkt_Fare', ascending=False) \
    [['Total_Mkt_Fare']]
Average_Mkt_Fare_arrive['id'] = Average_Mkt_Fare_arrive.index

Average_Mkt_Fare_arrive
```

```python
new_df2 = Average_Mkt_Fare_arrive.merge(citi, how='inner', on='id')

new_df2.iloc[:10]
```

```python
top10_arrive = new_df2.iloc[:10]
ax = sns.barplot(x='city', y='Total_Mkt_Fare', data=top10_arrive, color='green')

bars = ax.containers[0]
ax.bar_label(bars, labels=[f'${value:.2f}' for value in bars.datavalues])

ax.set_title('Top 10 Average Total Market Fare by Arrival City')

plt.show()
```
