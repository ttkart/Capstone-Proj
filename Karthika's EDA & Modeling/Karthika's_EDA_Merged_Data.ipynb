**Average Ticket Price by Departure city (Top 5)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate average ticket price for each departure city
avg_price_by_city = df.groupby('depart_city')['Total_Mkt_Fare'].mean().sort_values(ascending=False)

# Select top 5 departure cities
top5_cities = avg_price_by_city.head(5).index

# Filter the dataframe for the top 5 cities
df_top5 = df[df['depart_city'].isin(top5_cities)]

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='depart_city', y='Total_Mkt_Fare', data=df_top5, ci=None)
plt.title('Average Ticket Price for Top 5 Departure Cities')
plt.xlabel('Departure City')
plt.ylabel('Average Ticket Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```
